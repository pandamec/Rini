```julia
using CSV, DataFrames
using LsqFit
using Statistics
```

```julia
P = df[!,"Load (uN)"]*1e-6      # [N]
h = df[!,"Depth (nm)"]*1e-9     # [m]
```

```julia
# -----------------------------
# 2. Find maximum load point
# -----------------------------
imax = argmax(P)

Pmax = P[imax]
hmax = h[imax]

println("Pmax = ", Pmax)
println("hmax = ", hmax)

# -----------------------------
# 3. Extract unloading segment
# -----------------------------
# Take portion after max load
Punload = P[imax:end]
hunload = h[imax:end]
```

```julia
# Optional: use only top 50% of unloading
n = length(Punload)
range = 1:Int(round(0.6n))

Punload_fit = Punload[range]
hunload_fit = hunload[range]
```

```julia
fig = Figure(resolution = (1200, 800));

font=45
ax = Axis(fig[1,1],
    xlabel = L"Depth (nm)",
    ylabel = L"Force (uN)",
    xlabelsize = font,
    ylabelsize = font,
    xticklabelsize = font - 2,
    yticklabelsize = font - 2,
    xgridstyle = :dash,        # dashed grid
    ygridstyle = :dash,
    xgridvisible = false,
    ygridvisible = false,
    )

    Makie.scatter!(ax,hunload_fit,Punload_fit)
    fig
```

```julia
X = [hunload_fit ones(length(hunload_fit))]
coeffs = X \ Punload_fit

S = coeffs[1]   # slope = stiffness

println("Stiffness S = ", S, " N/m")
```

```julia
# -----------------------------
# 6. Contact depth
# -----------------------------
ε = 0.1
   # Berkovich
hc = hmax - ε * Pmax / S

println("Contact depth hc = ", hc)

# -----------------------------
# 7. Contact area
# -----------------------------
Ac = 24.5 * hc^2

println("Contact area Ac = ", Ac)

# -----------------------------
# 8. Reduced modulus
# -----------------------------
β = 1.05

Er = (1 / (2 * β)) * (S/ sqrt(pi*Ac))
```

```julia
# -----------------------------
# After computing Er (same as before)
# -----------------------------

# Film thickness
t = 10e-6   # 10 µm

# Doerner–Nix model
# -----------------------------
x = hc / t

Er_film = 1 / ( (1/Er) - x * ( (1/Ers) - (1/Er) ) )

println("Film reduced modulus = ", Er_film)

# -----------------------------
# Convert to elastic modulus
# -----------------------------
νf = 0.3   # parylene

Ei = 1140e9
νi = 0.07

Ef = (1 - νf^2) / ((1 / Er_film) - (1 - νi^2) / Ei)

println("Corrected film modulus Ef = ", Ef, " Pa")
```
