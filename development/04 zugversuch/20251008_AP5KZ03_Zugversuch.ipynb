## Functions

```julia

function import_TIRA(file_path)
    
    df=CSV.read(file_path,DataFrame) 
    df.Zeit = parse.(Float64, replace.(df.Zeit, "," => "."))
    df.Länge = parse.(Float64, replace.(df.Länge, "," => "."))
    df.Weg = parse.(Float64, replace.(df.Weg, "," => "."))
    #df.Dehnung = parse.(Float64, replace.(df.Dehnung, "," => "."))
    df.Kraft = parse.(Float64, replace.(df.Kraft, "," => "."))
    df.dL_ORG = parse.(Float64, replace.(df.dL_ORG, "," => "."))

    return df

end

function filter_range(df::DataFrame, column::String, low, high)
    return filter(row -> low ≤ row[column] ≤ high, df)
end
```

```julia
function get_corresponding(df::DataFrame, search_col::Symbol, search_val, return_col::Symbol)
    row = findfirst(df[!, search_col] .== search_val)
    return row === nothing ? missing : df[row, return_col]
end
```

```julia
function u_E(X,Y)
    p_fit=fit(X,Y,1)
    y_fit = p_fit.(X)
    residuals = y_fit - Y   # residuals
    
    σ_res = std(residuals, corrected=true) 
    X_mean = mean(X)
    stderr_slope = σ_res / sqrt(sum((X .- X_mean).^2))


    return stderr_slope
end
```

```julia
function getMechanicalProperties(fig,ax1,ax2,log_total,df_Group,LinearRange,StressRange,Temperature)

        log=[]
        for i in df_Group
            
            T=i[!,:Zeit]
            X=i[!,:dL_ORG]
            Y=i[!,:Stress]
            df_linear=filter_range(i,"Stress",StressRange[1],StressRange[2])
            df_linear=filter_range(df_linear,"dL_ORG",LinearRange[1],LinearRange[2])
            T_linear=df_linear[!,:Zeit]
            X_linear=df_linear[!,:dL_ORG]
            Y_linear=df_linear[!,:Stress]

            sigma0_25=maximum(Y_linear)
            sigmaB=maximum(Y)
            sigmaF=maximum(Y_linear)
            # Fit a 1st-degree polynomial (linear)
            fit_X_T=fit(T_linear,X_linear*60,1)
            
            strainRate=fit_X_T[1]
            
            # Fit a 1st-degree polynomial (linear)
            p_fit=fit(X_linear/100,Y_linear,1)
            E_modul=p_fit[1]/1000
            u_EModul=u_E(X_linear/100,Y_linear/1000)


            Makie.scatter!(ax1,X,Y; markersize=12, label = "$(round(strainRate, digits=3)) [%/min]") #Just for the paper
            #Makie.scatter!(ax1,X,Y; markersize=12,)
            Makie.scatter!(ax2,X_linear,Y_linear; markersize=12, label = "$(round(strainRate, digits=3)) [%/min]")
            
            δ_B=get_corresponding(i,:Stress,maximum(i[!,:Stress]),:dL_ORG)
            sum=Dict("Probe"=>i[1,:Name],
                    " Strain Rate[%/min]"=>strainRate,
                    "E[GPa] "=>E_modul,
                    "u_E[GPa] "=>u_EModul,
                    "sigma0_25% "=> sigma0_25,
                    "sigmaMax"=> sigmaB,
                    "δ_B"=> δ_B,
                    "δ_F/δ_B"=>df_linear[end,:dL_ORG]*100/δ_B,
                    "Temperature [C]"=> Temperature,
                    "LinearRange [%]"=>[df_linear[1,:dL_ORG],df_linear[end,:dL_ORG]])
            push!(log,sum)
            push!(log_total,sum)
            
        end

        axislegend(ax1, labelsize=font-10, framevisible=false)
        axislegend(ax2, labelsize=font-10, framevisible=false)

   

        df_log=DataFrame(log)
        return df_log,log_total,fig

end
```

```julia
function plotVariableTemperature!(fig,ax1,ax2,df,LinearRange,StressRange,Temperature)

        log=[]
        
            
            T=df[!,:Zeit]
            X=df[!,:dL_ORG]
            Y=df[!,:Stress]
            df_linear=filter_range(df,"Stress",StressRange[1],StressRange[2])
            df_linear=filter_range(df_linear,"dL_ORG",LinearRange[1],LinearRange[2])
            T_linear=df_linear[!,:Zeit]
            X_linear=df_linear[!,:dL_ORG]
            Y_linear=df_linear[!,:Stress]

            # Fit a 1st-degree polynomial (linear)
            fit_X_T=fit(T_linear,X_linear*60,1)
            
            strainRate=fit_X_T[1]


            Makie.scatter!(ax1,X,Y; markersize=12, label = "T=$Temperature C v=$(round(strainRate, digits=3)) [%/min]")
            Makie.scatter!(ax2,X_linear,Y_linear; markersize=12, label = "T=$Temperature C v=$(round(strainRate, digits=3)) [%/min]")
   
        return fig

end
```

```julia
function plotProperties!(fig, df_log,ax1,ax2,ax3)
        

        epsilon=df_log[!," Strain Rate[%/min]"]
        E=df_log[!,"E[GPa] "]
        Makie.scatter!(ax1,epsilon,E)
        Makie.lines!(ax1,epsilon,E)

        Makie.scatter!(ax2,df_log[!," Strain Rate[%/min]"],df_log[!,"sigmaMax"])
        Makie.lines!(ax2,df_log[!," Strain Rate[%/min]"],df_log[!,"sigmaMax"])


        Makie.scatter!(ax3,df_log[!," Strain Rate[%/min]"],df_log[!,"sigma0_25% "])
        Makie.lines!(ax3,df_log[!," Strain Rate[%/min]"],df_log[!,"sigma0_25% "])
        
        
        return fig

end
```
